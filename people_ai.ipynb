```python
# You are given a string s that contains some variables, along with a lookup table that contains key-value pairs.
# For example, in the string "(name)is(age)yearsold", there are two variables identified by the keys "name" and "age"
# You are also given a 2D lookup table with a wide range of keys. Each lookup[i] = [key_i, value_i] indicates that key key_i has a value of value_i.
# The task is to replace the variables with their respective value and return the new string
# If you do not know the value of the key, you will replace keyi with “?”

# Example 1:
#     Input: s = "(name)is(age)yearsold", lookup = [["name","bob"],["age","two"],["gender", "male"]]
#     Output: "bobistwoyearsold"
# Explanation:
#     The key "name" has a value of "bob", so replace "(name)" with "bob".
#     The key "age" has a value of "two", so replace "(age)" with "two".

# Example 2:
#     Input: s = "hi(name)", lookup = [["age","two"],["gender", "male"]]
#     Output: "hi?"
# Explanation:
#     The key "name" does not exist in lookup and is replaced with ?

# def get_input_info():
#     print("Lookup:")
#     print("Enter the look")
    


if __name__ == '__main__':
    # lookup, inp = get_input_info()
    lookup = [["name","bob"],["age","two"],["gender", "male"]]
    inp = "(qwerty)is(poiuy)yearsold"
    
    lookup_dict = {}
    
    # convert the lookup array to dict
    for val in lookup:
        lookup_dict[val[0]] = val[1]
    
    start_idx = inp.find('(')
    while start_idx != -1:
        end_idx = inp.find(')')
        strg_to_look = inp[start_idx+1:end_idx]
        
        # lookup_keys = 
        
        if strg_to_look in lookup_dict:
            modified_strg = inp[:start_idx]+lookup_dict[strg_to_look]+inp[end_idx+1:]
        else:
            modified_strg = inp[:start_idx]+'?'+inp[end_idx+1:]
        
        inp = modified_strg
        start_idx = inp.find('(')
    
    print(inp)
```
