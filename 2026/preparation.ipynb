1. Do arguments in Python get passed by reference or by value?
​
Python uses object reference semantics - mutable objects (lists, dicts) are passed by reference (changes affect original), immutable (ints, strings) by value. Example:

```python
def modify(lst):
    lst.append(4)  # Modifies original

my_list = [1,2,3]
modify(my_list)  # my_list = [1,2,3,4]

print(my_list)  # Output: [1, 2, 3, 4]
```

3. What are list/dict comprehensions? Give filter/reduce example.

```python
# List comp
squares = [x**2 for x in range(10) if x % 2 == 0]  # [0,4,16,36,64]
print(squares)

# Dict comp  
word_lengths = {word: len(word) for word in ['python','interview']}
print(word_lengths)

# Filter + reduce (functools.reduce)
from functools import reduce
evens_sum = reduce(lambda x,y: x+y, filter(lambda x: x%2==0, range(10)))  # 20
print(evens_sum)
```

4. Explain closures with Lambda.
​
Closure captures enclosing scope variables:

```python
def outer(x):
    return lambda y: x + y  # Captures 'x'
add5 = outer(5)
add5(3)  # 8
```
