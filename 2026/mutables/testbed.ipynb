```python
# mutable -> list, dict, set, bytearray
# immutable -> int, float, bool, str, tuple, frozenset, bytes, complex

a = 10
b = 10

print("a == b -> ", a == b)
print("a is b -> ", a is b)
```

```python
a = 257
b = 257

print("a == b -> ", a == b)
print("a is b -> ", a is b)
```

```python
s1 = "hello"
s2 = "hello"
print(s1 is s2)

s1 = "The quick brown fox jumps over the lazy dog"
s2 = "The quick brown fox jumps over the lazy dog"
print(s1 is s2)
```

```python
lst = [1, 2, 3]
lst.append(4)
print(lst)  # [1, 2, 3, 4]

tup = (1, 2, 3)
# tup[0] = 100  # ❌ Error! Tuples are immutable
```

```python
l1 = [1, 2, 3]
l2 = l1  # Both variables point to the same list in memory
l1.append(4)  # The same object is modified
print(l2)  # [1, 2, 3, 4] ✅
```

```python
l1 = [1, 2, 3]
l2 = l1
l1 = l1 + [4]  # A new list is created, and l1 now points to it
print(l2)  # [1, 2, 3] ❌ (l2 still holds the original list)
print(l1)
```

```python
def modify(x):
    print("Inside modify, before change: ", x)
    x = 100  # Creates a new object
    print("Inside modify, after change: ", x)

num = 50
print("Before modify: ", num)
modify(num)
print("After modify: ", num)  # 50 (unchanged)
```

```python
x = [1, 2, 3]
print(id(x))        # Outputs object's unique identifier (memory address)
print(type(x)) 
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)   # True, they have equal contents
print(a is b)   # False, they're different objects
```

```python
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)  # [1, 2, 3, 4]

before = id(numbers)
numbers.append(5)
after = id(numbers)
print(before == after)  # True
print(numbers) 
```

```python
x = 5
print(id(x))
x += 1              # This rebinds x to a new object
print(id(x)) 
```

```python
s = "hello"
print(id(s))
s += " world"
print(id(s))        # New object, new id
print(s)  
```

```python
# Mutable example
def modify_list(lst):
    lst.append(99)

l = [1, 2, 3]
modify_list(l)
print(l)  # [1, 2, 3, 99], because lists are mutable

# Immutable example
def modify_int(n):
    n += 1

x = 5
modify_int(x)
print(x)  # 5, unchanged
```

```python
def add_element(mylist):
    mylist.append('new')

lst = [0]
add_element(lst)
print(lst)  # [0, 'new']
```

```python
def add_one(num):
    num += 1
    print(f"Inside function: {num}")

n = 10
add_one(n)
print(f"Outside function: {n}")
```

```python
a = [1, 2, 3]
print(f"a: {a}") 

# Approach 1: Simple assignment (both refer to same object)
b = a
b.append(4)
print(f"a: {a}")   # [1, 2, 3, 4] - a was affected!

# Approach 2: Using .copy() (creates separate object)
c = a.copy()
c.append(5)
print(f"a: {a}")   # [1, 2, 3, 4] - unchanged
print(f"c: {c}")   # [1, 2, 3, 4, 5]

# Approach 3: Using slicing (also creates separate object)
d = a[:]
d.append(6)
print(f"a: {a}")   # [1, 2, 3, 4] - still unchanged
print(f"d: {d}")   # [1, 2, 3, 4, 6]
```

```python
# Small integers are cached
x = 256
y = 256
print(x is y)  # True - same object!

# Large integers are not
a = 1000
b = 1000
print(a is b)  # False - different objects
```
