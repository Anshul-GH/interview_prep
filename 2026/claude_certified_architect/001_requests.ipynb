# Create a working chatbot

```python
import os
import anthropic
from dotenv import load_dotenv

env_path = ".env"
load_dotenv(dotenv_path=env_path, override=True)

client = anthropic.Anthropic(
    # base_url="http://127.0.0.1:11434",
    # api_key="fake-key",
    base_url=os.getenv("ANTHROPIC_BASE_URL"),
    api_key=os.getenv("ANTHROPIC_API_KEY"),
)

llm_model = "llama3.2"

def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": llm_model,
        "max_tokens": 1024,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text

def start_chat_bot(system_prompt, temperature):
    # list of messages as history
    messages = []

    # run the chatbot forever

    while True:
        user_input = input("> ")
        print("> ", user_input)

        add_user_message(messages, user_input)

        answer = chat(messages, system_prompt, temperature)
        # answer = chat(messages)

        add_assistant_message(messages, answer)

        print(">> ", answer)
```

```python
import os
import anthropic
from dotenv import load_dotenv

env_path = ".env"
load_dotenv(dotenv_path=env_path, override=True)

client = anthropic.Anthropic(
    # base_url="http://127.0.0.1:11434",
    # api_key="fake-key",
    base_url=os.getenv("ANTHROPIC_BASE_URL"),
    api_key=os.getenv("ANTHROPIC_API_KEY"),
)

llm_model = "llama3.2"

messages = []

add_user_message(messages, "Writte a 1 sentence description of a fake dattabase")

with client.messages.stream(
    model=llm_model,
    max_tokens=1024,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")

stream.get_final_message()
```
