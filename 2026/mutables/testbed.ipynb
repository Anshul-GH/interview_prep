```python
l1 = [1, 2, 3]
l2 = l1  # Both variables point to the same list in memory
l1.append(4)  # The same object is modified
print(l2)  # [1, 2, 3, 4] ✅
```

```python
l1 = [1, 2, 3]
l2 = l1
l1 = l1 + [4]  # A new list is created, and l1 now points to it
print(l2)  # [1, 2, 3] ❌ (l2 still holds the original list)
print(l1)
```

```python
def modify(x):
    print("Inside modify, before change: ", x)
    x = 100  # Creates a new object
    print("Inside modify, after change: ", x)

num = 50
print("Before modify: ", num)
modify(num)
print("After modify: ", num)  # 50 (unchanged)
```

```python
# Mutable example
def modify_list(lst):
    lst.append(99)

l = [1, 2, 3]
modify_list(l)
print(l)  # [1, 2, 3, 99], because lists are mutable

# Immutable example
def modify_int(n):
    n += 1

x = 5
modify_int(x)
print(x)  # 5, unchanged
```

```python
def add_element(mylist):
    mylist.append('new')

lst = [0]
add_element(lst)
print(lst)  # [0, 'new']
```

```python
def add_one(num):
    num += 1
    print(f"Inside function: {num}")

n = 10
add_one(n)
print(f"Outside function: {n}")
```

```python
a = [1, 2, 3]
print(f"a: {a}") 

# Approach 1: Simple assignment (both refer to same object)
b = a
b.append(4)
print(f"a: {a}")   # [1, 2, 3, 4] - a was affected!

# Approach 2: Using .copy() (creates separate object)
c = a.copy()
c.append(5)
print(f"a: {a}")   # [1, 2, 3, 4] - unchanged
print(f"c: {c}")   # [1, 2, 3, 4, 5]

# Approach 3: Using slicing (also creates separate object)
d = a[:]
d.append(6)
print(f"a: {a}")   # [1, 2, 3, 4] - still unchanged
print(f"d: {d}")   # [1, 2, 3, 4, 6]
```

```python
# Small integers are cached
x = 256
y = 256
print(x is y)  # True - same object!

# Large integers are not
a = 1000
b = 1000
print(a is b)  # False - different objects
```

```python
def append_to_list(item, lst=[]):
    lst.append(item)
    return lst

print(append_to_list(1))
print(append_to_list(2))
```

```python
class A: 
    def foo(self):
        return "A"

class B(A):
    pass

class C(A):
    def foo(self):
        return "C"

class D(B, C):
    pass

a = A()
b = B()
c = C()
d = D()

print(a.foo())
print(b.foo())
print(c.foo())
print(d.foo())
```

```python
lst = [[ ]] * 3
lst[0].append(1)
print(lst)  # [[1], [1], [1]] ← All inner lists are SAME!
```

```python
lst = [1, 2]
a = lst; lst += [3]    # Mutates in place
print(a)  # [1, 2, 3]
print(lst)

lst = [1, 2] 
a = lst; lst = lst + [3]  # New list!
print(a)  # [1, 2]
print(lst)
```

```python
nums = [1, 2, 3, 4]
nums[1:3] = [10, 20]  # Replaces slice
print(nums)  # [1, 10, 20, 4]

nums[1:3] = [100]     # Different length OK
print(nums)  # [1, 100, 4]
```

```python
nums = [1, 2, 3, 4, 5]
print(nums[3:0:-1])  # [4, 3] ← Reversed slice!
```

```python
def count():
    yield 1
    yield 2

g = count()
print(next(g))  # 1
print(list(g))  # [2]
print(list(g))  # [] ← Dead!
```

```python
it = iter(range(3))
print(list(it))  # [0, 1, 2]
print(list(it))  # [] ← Exhausted!

# vs
print([x for x in range(3)])  # [0, 1, 2] ← Fresh each time
```
